```
FoS = MAG.getDataframe('FieldsOfStudy')

FoS.show(10)
FoS.createOrReplaceTempView('FoS')
```

```
PaperFoS = MAG.getDataframe('PaperFieldsOfStudy')

PaperFoS.show(10)
PaperFoS.createOrReplaceTempView('PaperFoS')
```

```
%sql 
SELECT PaperFoS.PaperId, PaperFoS.FieldOfStudyId, FoS.NormalizedName, FoS.PaperCount
FROM PaperFoS
LEFT JOIN FoS
ON PaperFoS.FieldOfStudyId = FoS.FieldOfStudyId
WHERE PaperId IN (
	SELECT _c2
	FROM df
)
```
